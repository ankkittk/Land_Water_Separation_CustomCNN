```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
import sys
sys.path.append(os.path.abspath('..'))
```

```python
from utils.preprocess import preprocess_pipeline
from cnn.cnn_model import CustomCNN
```

```python
path = "../data/coastline/3.png"
image = cv2.imread(path)
image = preprocess_pipeline(path)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
model = CustomCNN()
pooled_map = model.forward(image)
plt.imshow(pooled_map)
```

```python
plt.hist(pooled_map.flatten(), bins=50)
```
