```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
import sys
sys.path.append(os.path.abspath('..'))
```

```python
from utils.preprocess import preprocess_pipeline
from cnn.conv import convolve
from cnn.activation import relu
from cnn.pooling import max_pooling
```

```python
path = "../data/lake/3.png"
image = cv2.imread(path)
image = preprocess_pipeline(path)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
kernel = np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]
])
```

```python
feature_map = convolve(image, kernel)
activated_map = relu(feature_map)
pooled_map = max_pooling(activated_map)
plt.imshow(pooled_map)
```
